## Upload datset

```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("dgomonov/new-york-city-airbnb-open-data")

print("Path to dataset files:", path)

# List files in the downloaded directory
print("Files in the dataset directory:", os.listdir(path))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('/kaggle/input/new-york-city-airbnb-open-data/AB_NYC_2019.csv')
df.head(10)
```

## Check for missing values

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum()
print(missing[missing > 0])
```

## Dropping Rows and Cols

```python
df.dropna(subset=['name', 'host_name'], inplace=True) #dropping name and host_name

df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

df['last_review'] = df['last_review'].fillna('Never Reviewed')

df.isnull().sum()
```

# EDA (EXPLORATORY DATA ANALYSIS)

## Distribution of Price

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=False)
plt.xlim(0, 500)  # limit x-axis for better view
plt.title('Price Distribution')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.show()
```

## Room Type Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='room_type', order=df['room_type'].value_counts().index)
plt.title('Room Type Distribution')
plt.show()
```

## Listings per Neighborhood Group

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='neighbourhood_group', order=df['neighbourhood_group'].value_counts().index)
plt.title('Listings per Neighbourhood')
plt.show()
```

## Average price per Borough

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='neighbourhood_group', y='price', errorbar=None)
plt.title('Average Price per Borough')
plt.show()
```

## Correlation Analysis

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## Top 10 Neighborhoods by Number of Listings

```python
top_neighborhoods = df['neighbourhood'].value_counts().head(10)
print(top_neighborhoods)

plt.figure(figsize=(10,6))
sns.barplot(x=top_neighborhoods.values, y=top_neighborhoods.index)
plt.title('Top 10 Neighborhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.show()
```

##  Price Variation in Top Neighborhoods

```python
top_10 = df[df['neighbourhood'].isin(top_neighborhoods.index)]

plt.figure(figsize=(12,6))
sns.boxplot(data=top_10, x='neighbourhood', y='price')
plt.ylim(0, 500)
plt.title('Price Distribution in Top 10 Neighborhoods')
plt.xticks(rotation=45)
plt.show()
```

## Histogram of Availability Over the Year

```python
plt.figure(figsize=(10,6))
sns.histplot(df['availability_365'], bins=30)
plt.title('Distribution of Yearly Availability')
plt.xlabel('Availability (days)')
plt.ylabel('Number of Listings')
plt.show()
```

## Map Visualization with Folium

```python
import folium
from folium.plugins import MarkerCluster

# Center the map on NYC
map_nyc = folium.Map(location=[40.7128, -74.0060], zoom_start=11)
marker_cluster = MarkerCluster().add_to(map_nyc)

# Sample to avoid slow rendering
sample_df = df.sample(1000)

for index, row in sample_df.iterrows():
    folium.Marker([row['latitude'], row['longitude']],
                  popup=f"{row['name']} (${row['price']})").add_to(marker_cluster)

map_nyc
```

## Room Type vs Availability vs Price

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='room_type', y='availability_365', data=df)
plt.title('Availability Across Room Types')
plt.show()

plt.figure(figsize=(12,6))
sns.violinplot(x='room_type', y='price', data=df)
plt.ylim(0,500)
plt.title('Price Distribution by Room Type')
plt.show()
```

## Price vs Latitude/Longitude Heatmap

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='longitude', y='latitude', hue='price', data=df[df['price'] < 500], palette='viridis')
plt.title('Price Distribution by Location')
plt.show()
```
